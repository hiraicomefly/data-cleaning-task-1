```python
import pandas as pd
```

IMPORT DATA USING PANDAS FUNCTION "pd.read_csv" for csv file

```python
dt_import = pd.read_csv (r'D:\DS\Data_Cleaning_portfolio\data_cleaning_challenge.csv')
dt_import
```

DROP COLUMNS "Unnamed: 9","Unnamed: 10"

```python
drop_extra = dt_import.drop(columns=["Unnamed: 9","Unnamed: 10"])
drop_extra
```

DELETE NULL VALUES IN DATA FRAME

```python
drop_nan = drop_extra[drop_extra["Row Type"].notna()]
drop_nan
```

```python
column_values = []
iteration = 0 

for i in drop_nan["Row Type"]:
    if "first_name" in i:
        iteration = iteration + 1
    column_values.append(iteration)

column_values
```

```python
iter_cols = drop_nan
iter_cols["Iteration"] = column_values
iter_cols
```

```python
drop_extra_columns_name_row = iter_cols[iter_cols['Row Type'] != "Row Type"]
drop_extra_columns_name_row
```

```python
name_dataframe = drop_extra_columns_name_row[drop_extra_columns_name_row["Row Type"].str.contains("first_name")]
name_dataframe.drop(columns=["Speed1","Speed2","Electricity","Effort","Weight","Torque"], inplace=True)
name_dataframe.rename(columns={"Row Type":"First name", "Iter Number": "Last Name", "Power1":"Date"}, inplace = True)
name_dataframe['First name'] = name_dataframe["First name"].str[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]
name_dataframe
```

```python
no_name_df = drop_extra_columns_name_row[~drop_extra_columns_name_row["Row Type"].str.contains("first name")]
no_name_df
```

```python
pd.merge(left=name_dataframe, right = no_name_df, how="inner", on ="Iteration")
```
